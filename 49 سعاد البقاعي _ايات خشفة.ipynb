**الخلاصة:** البيانات متنوعة وغنية، ويجب تجهيز النصوص الطويلة لضمان أداء أفضل للنماذج.


```python
import re
import pandas as pd
from collections import Counter


GREETINGS = [
    'السلام عليكم ورحمة الله وبركاته',
    'السلام عليكم',
    'مرحبا', 'اهلا', 'تحية طيبة'
]

MEDICAL_TRIGGERS = [
    'اعاني', 'اعاني من', 'اشعر', 'اشعر ب',
    'اصبت', 'اصبت ب', 'لدي', 'ظهرت',
    'تناولت', 'اخذت', 'دواء', 'حبوب',
    'كريم', 'قطره', 'نقط',
    'الم', 'وجع', 'حكه', 'نزول',
    'انسداد', 'التهاب', 'ارتفاع'
]

def remove_greetings(text):
    for g in GREETINGS:
        text = text.replace(g, '')
    return text.strip()

def clean_arabic_text(text):
    if not isinstance(text, str):
        return ''


    text = re.sub(r'[ًٌٍَُِّْ]', '', text)


    text = re.sub('[إأآا]', 'ا', text)
    text = re.sub('ى', 'ي', text)
    text = re.sub('ؤ', 'و', text)
    text = re.sub('ئ', 'ي', text)
    text = re.sub('ة', 'ه', text)

    text = re.sub(r'[^؀-ۿ0-9\s]', ' ', text)


    text = re.sub(r'\s+', ' ', text).strip()

    return text


def clean_arabic_text(text):
    if not isinstance(text, str):
        return ''


    text = re.sub(r'[ًٌٍَُِّْ]', '', text)


    text = re.sub('[إأآا]', 'ا', text)
    text = re.sub('ى', 'ي', text)
    text = re.sub('ؤ', 'و', text)
    text = re.sub('ئ', 'ي', text)
    text = re.sub('ة', 'ه', text)


    text = re.sub(r'[^؀-ۿ0-9\s]', ' ', text)


    text = re.sub(r'\s+', ' ', text).strip()

    return text


def analyze_common_words(df, column='Question', top_n=30):
    all_words = ' '.join(df[column].dropna()).split()
    word_freq = Counter(all_words)
    most_common = word_freq.most_common(top_n)
    return word_freq, most_common

word_freq, most_common = analyze_common_words(df_train)


BASE_STOPWORDS = {
    'في','من','على','الى','عن','مع','او','ثم',
    'هذا','هذه','ذلك','الذي','التي','كان','كانت',
    'هل','لا','ما','لم','انا','لي','بعد','منذ',
    'عليكم','السلام','الله','رحمه','وبركاته'
}

COMMON_WORDS_STOP = set([w for w, _ in most_common])
ARABIC_STOPWORDS = BASE_STOPWORDS.union(COMMON_WORDS_STOP)



def remove_stopwords(text):
    return ' '.join([w for w in text.split() if w not in ARABIC_STOPWORDS])



def split_into_sentences(text, min_len=15):
    sentences = re.split(r'[.!؟،\n]', text)
    sentences = [s.strip() for s in sentences if len(s.strip()) >= min_len]
    return sentences


def extract_medical_core_from_sentences(text, max_sentences=3):
    sentences = split_into_sentences(text)
    selected = []

    for sent in sentences:
        if any(trig in sent for trig in MEDICAL_TRIGGERS):
            selected.append(sent)
        if len(selected) >= max_sentences:
            break

    return ' '.join(selected)


def full_preprocess(text, mode='full'):
    """
    mode = 'full' → نص نظيف كامل (AraBERT)
    mode = 'core' → نص مختصر ذكي للتصنيف (TF-IDF)
    """
    if not isinstance(text, str):
        return ''

    text = remove_greetings(text)
    text = clean_arabic_text(text)
    text = remove_stopwords(text)

    if mode == 'core':
        core = extract_medical_core_from_sentences(text)
        return core if len(core.split()) > 3 else text

    return text

for split in [df_train, df_val, df_test]:
    split['clean_full'] = split['Question'].apply(lambda x: full_preprocess(x, 'full'))
    split['clean_core'] = split['Question'].apply(lambda x: full_preprocess(x, 'core'))



def display_question_preprocessing(df, n=4):
    print(f"عرض {n} أمثلة من الأسئلة قبل وبعد المعالجة:\n")
    for i in range(n):
        question = df.iloc[i]['Question']
        clean_full = df.iloc[i]['clean_full']
        clean_core = df.iloc[i]['clean_core']

        print(f"مثال {i+1}:\n")
        print(f"Question:\n{question}\n")
        print(f"(clean_full):\n{clean_full}\n")
        print(f"(clean_core):\n{clean_core}\n")
        print("-" * 100)


display_question_preprocessing(df_train, n=4)
```

```python

# NLP HW1 - Medical Question Classification
# Method 2: Deep Learning (Embedding + BiLSTM)


!pip install -q datasets

from datasets import load_dataset
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Embedding, LSTM, Bidirectional, GlobalMaxPool1D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import regularizers



dataset = load_dataset("Ahmed-Selem/Shifaa_Arabic_Medical_Consultations")
df = dataset["train"].to_pandas()

df["Main_Category"] = df["Hierarchical Diagnosis"].apply(
    lambda x: x.split(" - ")[0] if isinstance(x, str) and " - " in x else x
)

df.rename(columns={"Question Text": "Question"}, inplace=True)
df = df.dropna(subset=["Question", "Main_Category"])



df_train, df_test = train_test_split(
    df, test_size=0.1, random_state=42, stratify=df["Main_Category"]
)
df_train, df_val = train_test_split(
    df_train, test_size=0.1, random_state=42, stratify=df_train["Main_Category"]
)



ARABIC_LETTERS = r"ءاأإآبتثجحخدذرزسشصضطظعغفقكلمنهوية"

def preprocess(text):
    text = str(text)
    text = re.sub("ّ|َ|ً|ُ|ٌ|ِ|ٍ|ْ|ـ", "", text)
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("[يى]", "ي", text)
    text = re.sub(f"[^{ARABIC_LETTERS}0-9 ؟!.,، ]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df_train["clean"] = df_train["Question"].apply(preprocess)
df_val["clean"]   = df_val["Question"].apply(preprocess)
df_test["clean"]  = df_test["Question"].apply(preprocess)



label_encoder = LabelEncoder()
df_train["label_id"] = label_encoder.fit_transform(df_train["Main_Category"])
df_val["label_id"]   = label_encoder.transform(df_val["Main_Category"])
df_test["label_id"]  = label_encoder.transform(df_test["Main_Category"])

num_classes = len(label_encoder.classes_)



tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(df_train["clean"])

vocab_size = len(tokenizer.word_index) + 1

X_train_seq = tokenizer.texts_to_sequences(df_train["clean"])
X_val_seq   = tokenizer.texts_to_sequences(df_val["clean"])
X_test_seq  = tokenizer.texts_to_sequences(df_test["clean"])

MAX_LEN = max(int(np.percentile([len(x) for x in X_train_seq], 95)), 15)

X_train = pad_sequences(X_train_seq, maxlen=MAX_LEN, padding="post")
X_val   = pad_sequences(X_val_seq,   maxlen=MAX_LEN, padding="post")
X_test  = pad_sequences(X_test_seq,  maxlen=MAX_LEN, padding="post")

y_train = df_train["label_id"].values
y_val   = df_val["label_id"].values
y_test  = df_test["label_id"].values



results_dict = {
    "number_step": [],
    "name_model": [],
    "features": [],
    "parameters_model": [],
    "methods_preprocessing": [],
    "accuracy": [],
    "F-score": []
}

def add_result(step, name, feat, params, prep, acc, f1):
    results_dict["number_step"].append(step)
    results_dict["name_model"].append(name)
    results_dict["features"].append(feat)
    results_dict["parameters_model"].append(params)
    results_dict["methods_preprocessing"].append(prep)
    results_dict["accuracy"].append(acc)
    results_dict["F-score"].append(f1)

prep_desc = "remove_diacritics, normalize_alef, normalize_ya, remove_non_arabic, remove_spaces"



EMB_DIM_RANDOM = 100

def build_random_model():
    inp = Input(shape=(MAX_LEN,))
    x = Embedding(vocab_size, EMB_DIM_RANDOM)(inp)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.5)(x)
    x = Dense(64, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = Dropout(0.5)(x)
    out = Dense(num_classes, activation="softmax")(x)
    model = Model(inp, out)
    model.compile(
        loss="sparse_categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy"]
    )
    return model

model1 = build_random_model()

early1 = EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)
ckpt1  = ModelCheckpoint("best_random_model.h5", save_best_only=True)

model1.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=12,
    batch_size=128,
    callbacks=[early1, ckpt1],
    verbose=1
)

model1.load_weights("best_random_model.h5")
y_pred1 = np.argmax(model1.predict(X_test), axis=1)

acc1 = accuracy_score(y_test, y_pred1)
f1_1 = f1_score(y_test, y_pred1, average="weighted")

add_result(
    step=1,
    name="Random_Embedding_BiLSTM",
    feat=f"seq_len={MAX_LEN}, emb_dim={EMB_DIM_RANDOM}",
    params="LSTM=64, Dense=64, dropout=0.5, l2=1e-4",
    prep=prep_desc,
    acc=acc1,
    f1=f1_1
)



!wget -q https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.ar.300.vec.gz
!gunzip -f cc.ar.300.vec.gz

EMB_DIM_PRE = 300
embeddings_index = {}

with open("cc.ar.300.vec", encoding="utf-8", errors="ignore") as f:
    for line in f:
        values = line.rstrip().split()
        if len(values) == EMB_DIM_PRE + 1:
            embeddings_index[values[0]] = np.asarray(values[1:], dtype="float32")

embedding_matrix = np.random.normal(0, 0.01, (vocab_size, EMB_DIM_PRE))
for word, idx in tokenizer.word_index.items():
    if word in embeddings_index:
        embedding_matrix[idx] = embeddings_index[word]

def build_pretrained_model():
    inp = Input(shape=(MAX_LEN,))
    x = Embedding(
        vocab_size,
        EMB_DIM_PRE,
        weights=[embedding_matrix],
        trainable=False
    )(inp)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.5)(x)
    x = Dense(64, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = Dropout(0.5)(x)
    out = Dense(num_classes, activation="softmax")(x)
    model = Model(inp, out)
    model.compile(
        loss="sparse_categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy"]
    )
    return model

model2 = build_pretrained_model()

early2 = EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)
ckpt2  = ModelCheckpoint("best_pretrained_model.h5", save_best_only=True)

model2.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=12,
    batch_size=128,
    callbacks=[early2, ckpt2],
    verbose=1
)

model2.load_weights("best_pretrained_model.h5")
y_pred2 = np.argmax(model2.predict(X_test), axis=1)

acc2 = accuracy_score(y_test, y_pred2)
f1_2 = f1_score(y_test, y_pred2, average="weighted")

add_result(
    step=2,
    name="Pretrained_Embedding_BiLSTM",
    feat=f"seq_len={MAX_LEN}, emb_dim={EMB_DIM_PRE}",
    params="LSTM=64, Dense=64, dropout=0.5, emb_trainable=False",
    prep=prep_desc,
    acc=acc2,
    f1=f1_2
)



results_df = pd.DataFrame(results_dict)
print("===== جدول مقارنة نماذج Deep Learning =====")
display(results_df)

best_model = results_df.loc[results_df["F-score"].idxmax()]

print("===== أفضل نموذج =====")
display(best_model)
```

نقوم باستخدام نموذج AraBERT مُسبق التدريب لاستفادة من المعرفة اللغوية السابقة باللغة العربية. عمليًا:
- نجهز البيانات (تنظيف الأسئلة واستخراج الفئة الرئيسية وترميزها).
- ندخل بياناتنا (الأسئلة والفئات) للنموذج، لكن بدل ما نسمح للنموذج يغيّر كل معرفته السابقة، نثبت (نجمد) بعض الطبقات الأولى من AraBERT لأنها تعلمت اللغة العربية مسبقًا. بهذه الطريقة، النموذج يستفيد من معرفته القديمة للغة، ونركز التدريب فقط على الطبقات العليا لتعليم النموذج كيفية تصنيف الأسئلة الجديدة.

- ندرب النموذج على تصنيف الأسئلة إلى فئاتها مع ضبط التعلم فقط على الطبقات العليا.
- نقيّم النموذج على بيانات الاختبار للحصول على دقة وأداء نهائي.


```python

# NLP HW1 - Method 3: Transfer Learning (AraBERT)
# Final Corrected & Organized Version

!pip install -q datasets transformers arabert


import torch
import torch.nn as nn
from torch.optim import AdamW
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np
import re

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    get_linear_schedule_with_warmup
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)



dataset = load_dataset("Ahmed-Selem/Shifaa_Arabic_Medical_Consultations")
df = dataset["train"].to_pandas()

df["Main_Category"] = df["Hierarchical Diagnosis"].apply(
    lambda x: x.split(" - ")[0] if isinstance(x, str) and " - " in x else x
)

df.rename(columns={"Question Text": "Question"}, inplace=True)
df = df.dropna(subset=["Question", "Main_Category"])

print("Dataset shape:", df.shape)



label_encoder = LabelEncoder()
df["label_id"] = label_encoder.fit_transform(df["Main_Category"])
num_labels = len(label_encoder.classes_)

train_df, test_df = train_test_split(
    df, test_size=0.1, random_state=42, stratify=df["label_id"]
)
train_df, val_df = train_test_split(
    train_df, test_size=0.1, random_state=42, stratify=train_df["label_id"]
)

print("Train:", train_df.shape, "Val:", val_df.shape, "Test:", test_df.shape)

y_train = train_df["label_id"].values



MODEL_NAME = "aubmindlab/bert-base-arabertv2"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels
)
model.to(device)

# Freeze first 6 BERT layers
for name, param in model.bert.named_parameters():
    if "encoder.layer" in name:
        layer_num = int(name.split("encoder.layer.")[1].split(".")[0])
        if layer_num < 6:
            param.requires_grad = False



MAX_LEN = 128
BATCH_SIZE = 16
EPOCHS = 6
LR = 2e-5

class ShifaaDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts.tolist()
        self.labels = labels.tolist()
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            self.texts[idx],
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )
        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = ShifaaDataset(train_df["Question"], train_df["label_id"], tokenizer, MAX_LEN)
val_dataset   = ShifaaDataset(val_df["Question"],   val_df["label_id"],   tokenizer, MAX_LEN)
test_dataset  = ShifaaDataset(test_df["Question"],  test_df["label_id"],  tokenizer, MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)



class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

loss_fn = nn.CrossEntropyLoss(weight=class_weights)

optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=LR)

total_steps = len(train_loader) * EPOCHS
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.1 * total_steps),
    num_training_steps=total_steps
)



def train_one_epoch(model, loader):
    model.train()
    total_loss, correct = 0, 0

    for batch in loader:
        optimizer.zero_grad()
        input_ids = batch["input_ids"].to(device)
        mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=mask)
        loss = loss_fn(outputs.logits, labels)

        loss.backward()
        optimizer.step()
        scheduler.step()

        preds = outputs.logits.argmax(dim=1)
        total_loss += loss.item()
        correct += (preds == labels).sum().item()

    return total_loss / len(loader), correct / len(train_df)


def eval_model(model, loader):
    model.eval()
    preds_all, labels_all = [], []

    with torch.no_grad():
        for batch in loader:
            input_ids = batch["input_ids"].to(device)
            mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=mask)
            preds = outputs.logits.argmax(dim=1)

            preds_all.extend(preds.cpu().numpy())
            labels_all.extend(labels.cpu().numpy())

    acc = accuracy_score(labels_all, preds_all)
    f1  = f1_score(labels_all, preds_all, average="weighted")
    return acc, f1, labels_all, preds_all



best_f1 = 0
best_state = None

for epoch in range(EPOCHS):
    train_loss, train_acc = train_one_epoch(model, train_loader)
    val_acc, val_f1, _, _ = eval_model(model, val_loader)

    print(f"Epoch {epoch+1}/{EPOCHS}")
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val Acc: {val_acc:.4f} | Val F1: {val_f1:.4f}")

    if val_f1 > best_f1:
        best_f1 = val_f1
        best_state = model.state_dict()


model.load_state_dict(best_state)



test_acc, test_f1, y_true, y_pred = eval_model(model, test_loader)

print("\n===== Final Test Results =====")
print("Accuracy:", test_acc)
print("F1-score:", test_f1)
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_, zero_division=0))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, cmap="Blues", cbar=False)
plt.title("Confusion Matrix - AraBERT")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()



results_dict = {
    "number_step": [3],
    "name_model": ["AraBERT (Transfer Learning)"],
    "features": ["AraBERT tokenizer, max_len=128"],
    "parameters_model": ["LR=2e-5, epochs=6, batch=16, freeze<6, class_weights"],
    "methods_preprocessing": ["AraBERT built-in tokenization"],
    "accuracy": [test_acc],
    "F-score": [test_f1]
}

results_df_method3 = pd.DataFrame(results_dict)

print("\n===== Results Table (Method 3) =====")
display(results_df_method3)
```
